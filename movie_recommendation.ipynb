```python
import numpy as np
import pandas as pd
```

```python
movies = pd.read_csv("tmdb_5000_movies.csv")
credits = pd.read_csv("tmdb_5000_credits.csv")
```

```python
# will merge the tables
movies=movies.merge(credits,on="title")
```

```python
movies.head(1)
```

```python
# we will make recommendation on basis of genres,keywords,cast(top3)&overview

movies=movies[['movie_id','title','genres','keywords','cast','overview']]
```

```python
movies.head(1)
```

```python
# data prosecing
movies.isnull().sum()
```

```python
movies.dropna(inplace=True)
```

```python
movies.duplicated().sum()
```

```python
movies.iloc[0].genres
```

```python
movies.iloc[0].keywords
```

```python
# genres & keywords are in form of string in list in dictionary we have to convert it into list
import ast
def convert(a):
    l=[]
    for i in ast.literal_eval(a):
        l.append(i['name'])
    return l
```

```python
movies['genres']=movies['genres'].apply(convert)
```

```python
movies.head(1)
```

```python
movies['keywords']=movies['keywords'].apply(convert)
```

```python
movies.head(1)
```

```python
# for cast i will take only top 3 cast name
movies['cast'][0]
```

```python
def convert2(a):
    l=[]
    counter=0
    for i in ast.literal_eval(a):
        if counter !=3:
         l.append(i['name'])
         counter+=1
        else:
            break
    return l
```

```python
movies['cast']=movies['cast'].apply(convert2)
```

```python
movies.head(1)
```

```python
movies['genres']=movies['genres'].apply(lambda x: " ".join([i.replace(" ", "")for i in x]))
movies['keywords']=movies['keywords'].apply(lambda x: " ".join([i.replace(" ", "")for i in x]))
movies['cast']=movies['cast'].apply(lambda x: " ".join([i.replace(" ", "")for i in x]))
```

```python
movies.head()
```

```python
movies['tags']=(
    movies['overview']+" "+movies['genres']+" "+movies['keywords']+" "+movies['cast']
).str.lower()
```

```python
movies.head()
```

```python
movies['tags'][1]
```

```python
movies=movies[['movie_id','title','tags']]
```

```python
movies.head()
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer

ps=PorterStemmer()
```

```python
def stem(text):
    y=[]
    for i in text.split():
       y.append(ps.stem(i))
    return " ".join(y)
```

```python
movies['tags'] = movies['tags'].apply(stem)
```

```python
movies['tags'][0]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
cv=CountVectorizer(max_features=2000,stop_words='english')
vectors=cv.fit_transform(movies['tags']).toarray()
```

```python
vectors.shape
```

```python
similarity=cosine_similarity(vectors)
```

```python
similarity.shape
```

```python
movies.head(1)
```

```python
def recommend(movie):
    movie_index=movies[movies['title']==movie].index[0]
    distances=similarity[movie_index]
    movies_list=sorted(list(enumerate(distances)),reverse=True,key= lambda x:x[1])[1:4]
    for i in movies_list:
        print(movies.iloc[i[0]].title)
```

```python
recommend
```
